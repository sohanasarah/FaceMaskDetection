```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
for dirname, _, filenames in os.walk('data/no_mask'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
image_size=100
image1 = cv2.imread('data/no_mask/1157.png')
image_gray= cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image_gray_resize= cv2.resize(image_gray,(image_size,image_size))
plt.imshow(image_gray_resize, cmap="gray")
plt.show()
```
